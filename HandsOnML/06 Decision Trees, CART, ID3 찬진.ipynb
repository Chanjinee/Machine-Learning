## 6. Decision Tree Algorithm

## Decision Tree란?
- 특정 질문을 기준으로 샘플들을 나누어 분류, 예측하는 알고리즘으로 나무가 가지를 뻗듯 질문을 거듭하며 뻗어가는 알고리즘
- Decision Tree는 데이터의 __전처리가 거의 필요하지 않음__ (스케일링 X)
- 알고리즘 특성상 __과적합에 취약함__

    - 주요 용어

|용어|뜻|
|:--------------:|-----------------|
|루트노드|'뿌리'노드, __맨꼭대기__ 노드|
|리프노드|'잎'노드, __맨 마지막__ 잎사귀와 같은 노드|
|불순도  |(<->순수도) 원하는 것 외에 다른 것이 섞여있는 정도/지니계수, 엔트로피

---

## CART Algorithm
- 불순도를 측정하는 지표로 지니계수를 사용
- 리프노드 외의 모든 노드가 자식노드를 두개씩 갖는 Binary Split
- 데이터가 가진 각 속성에 대해 분할 후 지니계수를 하여 지니계수가 가장 낮은 속성으로 샘플들을 분할해 나가는 방식

    - 지니계수?
    
$$ G(S) = 1-\sum_{i=1}^c p_i^2 $$



S : 이미 발생한 사건의 모음, c : 사건의 갯수
- 지니계수는 0 ~ 0.5의 값을 가짐
- 지니계수가 0이라는 것은 불순도가 0. 즉, 완전히 순수하다는 뜻

---

## ID3 Algorithm
- 불순도를 측정하는 지표로 엔트로피를 사용
- CART와 다르게 자식노드를 3개 이상 가질 수 있음
- 독립변수가 모두 범주형인 경우에만 사용이 가능하다는 단점이 있음  
(연속형 변수도 가능하도록 한 것은 C4.5)
- 모든 속성에 대하여 분할을 진행해 보고 정보획득량이 많은, 엔트로피가 작은 속성을 기준으로 분할

엔트로피와 정보 획득량

    - 엔트로피?

- 엔트로피는 정보량의 기댓값. 즉, 엔트로피가 크다는 것은 정보량이 많아 불순도가 높음

    - 정보획득량?
- 분할전 엔트로피와 분할후 엔트로피의 차이

---

## 독립변수가 연속형인 경우
    - 독립변수가 연속형일 때, 분할하는 방법 -> 무엇을 기준으로 분할할 것인가?
- 전체데이터를 모두 기준점으로
- 중위수, 사분위수를 기준점으로
- Label의 Class가 바뀌는 지점을 기준점으로

이후의 과정은 CART방식과 동일

---

## 가지치기
>사전 가지치기(pre-pruning)
>>조기 정지 규칙에 의해 트리 성장을 멈춘다.
정지 규칙으로는 트리의 깊이, 마지막 노드의 최소 데이터 수, 불순척도 등의 임계치를 이용하는 방법이 있다.

>사후 가지치기(post-pruning)
>>초기에는 트리를 최대의 크기로 만든다. 마지막 노드의 불순 척도가 최소가 되도록 분할한다. 복잡도가 최대가 된다.
Cross Validation (CV) 시험 오차가 최소가 되는 분할 수준으로 트리를 줄인다.

## Regression
- Classification과의 주된 차이점은 Class가 아닌 어떤 값을 예측한다는 것
- 마지막 리프노드에 있는 샘플들의 평균 타겟값이 예측값이 되는 방식
- 불순도를 최소화하는 방향이 아닌 MSE를 최소화하는 방향으로 학습됨

## Code 

### 타이타닉 데이터에 적용

```python
# 모듈 불러오기
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from tqdm import tqdm_notebook
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier

# Waeing
import sys, warnings
if not sys.warnoptions: warnings.simplefilter("ignore")
    
#matplotlib 한글깨짐 지원
import platform
from matplotlib import font_manager, rc
path = "c:/Windows/Fonts/malgun.ttf"
if platform.system() == 'Darwin':
    rc('font', family='AppleGothic')
elif platform.system() == 'Windows':
    font_name = font_manager.FontProperties(fname=path).get_name()
    rc('font', family=font_name)
else:
    print('Unknown system...')
rc('axes', unicode_minus=False)  
```

- 데이터 불러오기

```python
# 데이터 불러오기
path = './datasets/titanic/'
train = pd.read_csv(path+'train.csv', index_col='PassengerId') # `PassengerId` 열을 인덱스 열로 지정
test = pd.read_csv(path+'test.csv', index_col='PassengerId')
```

```python
train
```

- 간단한 전처리

```python
# encoder 함수 지정
def encoder(data, kind):
    # 범주형 변수와 수치형 변수를 분리
    cat_features = data.select_dtypes(include=['object']).columns.to_list()
    num_features = data.select_dtypes(exclude='object').columns.to_list() 
    if kind == 'onehot':
        # 원핫인코딩
        data_cat = pd.get_dummies(data[cat_features])
        data = pd.concat([data,data_cat],axis=1).drop(columns=[f for f in cat_features])
        return data
    elif kind == 'label':
        # 라벨인코딩
        for c in cat_features:
            data[c]=pd.Categorical(data[c]).codes
        return data
    else:
        print('choose "onehot"or "label"')
# 인코딩
train = encoder(train, 'label')
test = encoder(test, 'label')
```

```python
# 간단한 결측값 처리
## train
train['Age'] = train['Age'].fillna(train['Age'].mode()[0])
## test
test['Age'] = test['Age'].fillna(test['Age'].mode()[0])
test['Fare'] = test['Fare'].fillna(test['Fare'].mode()[0])
```

```python
# 데이터를 학습과 검증으로 분리
x_train, x_valid, y_train, y_valid = train_test_split(train.iloc[:,1:], train.iloc[:,0], test_size=0.2)
```

- 지니계수 vs 엔트로피

```python
# 학습의 결과를 눈으로 보기위해 score를 받아줌
trainGini = []
testGini = []
trainEntropy = []
testEntropy = []

for k in tqdm_notebook(range(1, 30)):
    # DT으로 train 데이터를 학습 (지니계수 사용)
    dt = DecisionTreeClassifier(criterion='gini', max_depth=k)
    dt.fit(x_train, y_train)
    
    # 정확도 측정
    trainGini.append(dt.score(x_train, y_train))
    testGini.append(dt.score(x_valid, y_valid))
    
    # DT으로 train 데이터를 학습 (엔트로피 사용)
    dt = DecisionTreeClassifier(criterion='entropy', max_depth=k)
    dt.fit(x_train, y_train)
    
    # 정확도 측정
    trainEntropy.append(dt.score(x_train, y_train))
    testEntropy.append(dt.score(x_valid, y_valid))
    
# 시각화하여 확인
plt.figure(figsize=(15, 8))
plt.plot(trainGini, label='trainGini')
plt.plot(trainEntropy, label='trainEntropy')
plt.plot(testGini, label='testGini')
plt.plot(testEntropy, label='testEntropy')
plt.legend()
plt.xlabel('Depth')
plt.ylabel('Accuracy')
plt.show()
```

- 사전 가지치기(pre-pruning)

```python
trainGini = []
testGini = []
trainEntropy = []
testEntropy = []
depth = []

# 사전 가지치기
for k in tqdm_notebook(range(1, 20)):
    dt = DecisionTreeClassifier(criterion='gini', max_depth=k)
    dt.fit(x_train, y_train)
    
    trainGini.append(dt.score(x_train, y_train))
    testGini.append(dt.score(x_valid, y_valid))
    
    dt = DecisionTreeClassifier(criterion='entropy', max_depth=k)
    dt.fit(x_train, y_train)
    
    trainEntropy.append(dt.score(x_train, y_train))
    testEntropy.append(dt.score(x_valid, y_valid))
    
    depth.append(k)
    print('depth = %d done.' % k)

plt.figure(figsize=(15, 8))
plt.plot(trainGini, label='trainGini')
plt.plot(trainEntropy, label='trainEntropy')
plt.plot(testGini, label='testGini')
plt.plot(testEntropy, label='testEntropy')
plt.legend()
plt.xlabel('Depth')
plt.ylabel('Accuracy')
plt.show()

# 정확도가 가장 큰 최적의 depth를 찾는다
nDepth = depth[np.argmax(testGini)]

# opt_alpha를 적용한 tree를 사용한다.
dt = DecisionTreeClassifier(max_depth = nDepth)
dt.fit(x_train, y_train)
print('시험 데이터의 정확도 = %.4f' % dt.score(x_valid, y_valid))
print('최적 트리의 depth = %d' % nDepth)
```

- 사후 가지치기(post-pruning)

```python
# 사후 가지치기
path = DecisionTreeClassifier().cost_complexity_pruning_path(x_train, y_train)
# alpha값과 error값을 리턴해줌
ccp_alphas, impurities = path.ccp_alphas, path.impurities # impurities: 불순도

plt.figure(figsize=(8,4))
plt.plot(ccp_alphas[:-1], impurities[:-1], marker='o', drawstyle='steps-post')
plt.xlabel('effective alpha')
plt.ylabel('total impurity of leaves')
plt.text(0.005, 0.15, 'alpha가 증가할수록 불순도가 증가한다.', fontsize=15)
plt.title('Total impurity vs effective alpha for training set')

# ccp_alphas가 너무 작은 것은 제외한다
ccp_alphas = ccp_alphas[np.where(ccp_alphas>0.0001)]

clfs = []
for i, ccp_alpha in enumerate(ccp_alphas):
    clf = DecisionTreeClassifier(ccp_alpha=ccp_alpha)
    clf.fit(x_train, y_train)
    clfs.append(clf)

print('마지막 tree의 노드 개수  = %d' % clfs[-1].tree_.node_count)
print('마지막 tree의 alpha  = %.4f' % ccp_alphas[-1])
print('마지막 tree의 depth  = %d' % clfs[-1].tree_.max_depth)

# 마지막 tree는 depth = 0이므로 제외한다.
clfs = clfs[:-1]
ccp_alphas = ccp_alphas[:-1]

# ccp_alphas가 증가할수록 node개수와 depth가 감소하는 것을 확인한다.
node_counts = [clf.tree_.node_count for clf in clfs]
depth = [clf.tree_.max_depth for clf in clfs]

plt.figure(figsize=(8,4))
plt.plot(ccp_alphas, node_counts, marker='o')
plt.xlabel('ccp_alphas')
plt.ylabel('node_counts')
plt.text(0.005, 150, 'alpha가 증가할수록 leaf-node의 개수가 감소한다.\n Tree가 단순해진다(underfitting).', fontsize=15)
plt.show()

plt.figure(figsize=(8,4))
plt.plot(ccp_alphas, depth, marker='o')
plt.xlabel('ccp_alphas')
plt.ylabel('depth')
plt.text(0.005, 10, 'alpha가 증가할수록 \ntree의 depth가 얕아진다(underfitting).', fontsize=15)
plt.show()

# ccp_alphas를 적용한 tree들 (clfs)로 score를 계산한다.
# clfs는 앞 부분 n개만 사용한다. 뒷 부분은 alpha가 너무 크기 때문에 제외한다.
n=70
train_scores = [clf.score(x_train, y_train) for clf in clfs[:n]]
test_scores = [clf.score(x_valid, y_valid) for clf in clfs[:n]]

plt.figure(figsize=(8,5))
plt.plot(ccp_alphas[:n], train_scores[:n], marker='o', label='trainAcc')
plt.plot(ccp_alphas[:n], test_scores[:n], marker='o', label='testAcc')
plt.xlabel('ccp_alphas')
plt.ylabel('score')
plt.legend()
plt.show()

# test_scores[:n]중 가장 큰 최적 alpha를 찾는다.
opt_alpha = ccp_alphas[np.argmax(test_scores[:n])]

# opt_alpha를 적용한 tree를 사용한다.
dt = DecisionTreeClassifier(ccp_alpha=opt_alpha)
dt.fit(x_train, y_train)

print('시험 데이터의 정확도 = %.4f' % dt.score(x_valid, y_valid))
print('Optimal alpha = %.8f' % opt_alpha)
```

- Feature 중요도 분석

```python
# 변수의 중요도를 시각적으로 확인
feature_importance = dt.feature_importances_
feature_name = list(train.columns)[1:] # target값 지워줌
n_feature = x_train.shape[1]
idx = np.arange(n_feature)

plt.bar(idx,feature_importance)
plt.xticks(rotation=90) #x축 값 세로로 변경
plt.xticks(idx, feature_name, fontsize=15)
plt.xlabel('feature', size=15)
plt.ylabel('feature importance', size=15)
plt.show()
```
